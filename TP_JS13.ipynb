# **Tugas Praktikum**

## **Model Dasar (2 Hidden Layer)**

```python
import tensorflow as tf
import time
import pandas as pd
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
```

```python
# 1. Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 -> 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Bangun model JST (Model Dasar)
model_base = Sequential([
    Flatten(input_shape=(28, 28)),    # Input Layer: Ubah gambar 28x28 jadi vektor 784
    Dense(128, activation='relu'),    # Hidden Layer 1
    Dense(64, activation='relu'),     # Hidden Layer 2
    Dense(10, activation='softmax')   # Output Layer (10 kelas angka 0-9)
])

# 3. Kompilasi model
model_base.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

# 4. Latih model
print("--- Training Model Dasar ---")
start_time = time.time()
history_base = model_base.fit(X_train, y_train, epochs=5, batch_size=32, verbose=1)
time_base = time.time() - start_time

# 5. Evaluasi model
loss_base, acc_base = model_base.evaluate(X_test, y_test, verbose=0)
print(f"Waktu Training: {time_base:.2f} detik")
print(f"Akurasi pada data uji: {acc_base:.4f}")
```

* Training Process: Output menampilkan proses pelatihan selama 5 epoch. Nilai loss akan menurun dan accuracy akan meningkat seiring berjalannya epoch.

* Waktu Training: Menunjukkan durasi yang dibutuhkan untuk melatih model dasar.

* Akurasi Akhir: Dataset MNIST adalah dataset standar yang relatif bersih, sehingga model sederhana dengan 2 hidden layer biasanya sudah mampu mencapai akurasi di atas 97% (0.97xx).

## **Eksperimen Modifikasi Arsitektur (Neuron Lebih Banyak + Layer Tambahan)**

Pengubahan jumlah neuron menjadi 256 dan 128.

Penambahan satu hidden layer lagi (total 3 hidden layer).

Penggunaan fungsi aktivasi ReLU.

```python
# Model Eksperimen 1: Lebih Dalam & Lebih Lebar (ReLU)
model_deep = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(256, activation='relu'),    # Hidden Layer 1 (Diperbesar)
    Dense(128, activation='relu'),    # Hidden Layer 2 (Diperbesar)
    Dense(64, activation='relu'),     # Hidden Layer 3 (Tambahan)
    Dense(10, activation='softmax')
])

model_deep.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("\n--- Training Model Deep (ReLU) ---")
start_time = time.time()
history_deep = model_deep.fit(X_train, y_train, epochs=5, batch_size=32, verbose=1)
time_deep = time.time() - start_time

loss_deep, acc_deep = model_deep.evaluate(X_test, y_test, verbose=0)
print(f"Waktu Training: {time_deep:.2f} detik")
print(f"Akurasi Model Deep: {acc_deep:.4f}")
```

* Waktu Training: Waktu yang dibutuhkan kemungkinan sedikit lebih lama dibandingkan model dasar karena jumlah parameter (bobot) yang harus dihitung lebih banyak (akibat penambahan neuron dan layer).

* Akurasi: Penambahan kapasitas model (neuron dan layer) pada dataset MNIST seringkali meningkatkan akurasi sedikit lebih tinggi dibanding model dasar, mendekati 98% atau lebih, karena model mampu menangkap fitur tulisan tangan yang lebih kompleks.

## **Eksperimen Fungsi Aktivasi Sigmoid**

```python
# Model Eksperimen 2: Arsitektur Sama, Aktivasi Sigmoid
model_sigmoid = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(256, activation='sigmoid'), # Ganti ReLU dengan Sigmoid
    Dense(128, activation='sigmoid'),
    Dense(64, activation='sigmoid'),
    Dense(10, activation='softmax')
])

model_sigmoid.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("\n--- Training Model Deep (Sigmoid) ---")
start_time = time.time()
history_sigmoid = model_sigmoid.fit(X_train, y_train, epochs=5, batch_size=32, verbose=1)
time_sigmoid = time.time() - start_time

loss_sigmoid, acc_sigmoid = model_sigmoid.evaluate(X_test, y_test, verbose=0)
print(f"Waktu Training: {time_sigmoid:.2f} detik")
print(f"Akurasi Model Sigmoid: {acc_sigmoid:.4f}")
```

* Performa Konvergensi: Model dengan Sigmoid biasanya menunjukkan peningkatan akurasi yang lebih lambat dibandingkan ReLU dalam jumlah epoch yang sama.

* Akurasi Akhir: Akurasi model Sigmoid kemungkinan lebih rendah dibandingkan model ReLU. Hal ini disebabkan oleh masalah vanishing gradient pada fungsi Sigmoid, di mana gradien menjadi sangat kecil pada layer-layer awal, menyulitkan model yang dalam (deep) untuk belajar secara efektif.

## **Perbandingan Hasil**

```python
results = {
    "Model": ["Dasar (2 Layer, Kecil)", "Deep (3 Layer, Besar, ReLU)", "Deep (3 Layer, Besar, Sigmoid)"],
    "Akurasi": [acc_base, acc_deep, acc_sigmoid],
    "Waktu (detik)": [time_base, time_deep, time_sigmoid]
}

df_results = pd.DataFrame(results)
print("\n--- Tabel Perbandingan ---")
print(df_results)
```

3. Model Sigmoid cenderung memiliki performa terburuk dalam hal akurasi pada arsitektur yang dalam (deep architecture) dibandingkan ReLU, membuktikan mengapa ReLU menjadi standar industri untuk hidden layer.
