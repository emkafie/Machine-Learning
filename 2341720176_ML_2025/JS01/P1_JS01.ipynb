Penjelasan Kegunaan masing-masing library tersebut:

1. PyPREP

Digunakan untuk pra-pemrosesan sinyal EEG (Electroencephalography).

Fungsinya menghilangkan noise, artefak (seperti kedipan mata, gerakan otot, atau gangguan listrik), dan menyiapkan data EEG biar lebih bersih sebelum analisis.

2. SciPy

Library Python untuk komputasi ilmiah dan teknik.

Berisi banyak modul matematika: aljabar linear, optimasi, integrasi, statistik, pemrosesan sinyal, dan banyak lagi.

3. Weights & Biases (wandb)

Platform untuk tracking eksperimen Machine Learning.

Dipakai buat mencatat hasil training model, menyimpan hyperparameter, visualisasi metrik (loss, akurasi), dan membandingkan berbagai percobaan.

4. pyECG

Library Python untuk menganalisis sinyal ECG (Electrocardiography).

Bisa membaca data ECG, ekstraksi fitur (misalnya deteksi puncak R, interval RR), hingga klasifikasi sinyal jantung.
