**1. Contoh Tindakan Melanggar Etika dan Hukum dalam Penggunaan Kecerdasan Buatan**

a. Deepfake untuk penyebaran hoaks dan pornografi non-konsensual

Penjelasan: Teknologi AI seperti deep learning dapat digunakan untuk membuat video atau audio palsu yang menyerupai orang lain. Ini berbahaya karena bisa digunakan untuk pencemaran nama baik, manipulasi politik, atau eksploitasi seksual.

Aspek hukum: Banyak negara menganggap tindakan ini melanggar hukum, misalnya terkait perlindungan data pribadi, UU ITE (di Indonesia), serta hak atas citra dan privasi individu.

Referensi:

Chesney, R., & Citron, D. (2019). Deep Fakes: A Looming Challenge for Privacy, Democracy, and National Security. California Law Review.

UU ITE No. 11 Tahun 2008 (Indonesia).

b. Diskriminasi dalam sistem rekrutmen berbasis AI

Penjelasan: AI yang digunakan untuk seleksi kerja bisa bias jika datanya tidak seimbang, misalnya lebih banyak meloloskan kandidat laki-laki daripada perempuan karena bias historis pada data.

Aspek hukum & etika: Melanggar prinsip keadilan dan non-diskriminasi dalam perekrutan tenaga kerja.

Referensi:

Barocas, S., Hardt, M., & Narayanan, A. (2019). Fairness and Machine Learning.

**2. Dampak Energi dan Lingkungan terhadap Pemanfaatan Kecerdasan Buatan**

a. Dampak Negatif

Konsumsi energi besar: Training model AI skala besar (seperti GPT atau BERT) membutuhkan server berdaya tinggi. Studi menyebutkan melatih satu model NLP besar bisa menghasilkan emisi karbon setara dengan lima kali perjalanan pulang-pergi pesawat antara New York dan San Francisco.

Peningkatan jejak karbon: Data center untuk AI menyumbang jejak karbon karena konsumsi listrik tinggi, apalagi jika listrik masih berasal dari energi fosil.

Referensi:

Strubell, E., Ganesh, A., & McCallum, A. (2019). Energy and Policy Considerations for Deep Learning in NLP.

b. Cara Mengatasinya (Pendapat Pribadi + Rekomendasi)

Optimasi model AI: Menggunakan teknik model compression, pruning, dan knowledge distillation agar model tetap efektif tapi tidak boros energi.

Green Data Center: Beralih ke server dengan energi terbarukan (solar, angin, hydro). Google dan Microsoft sudah memulai inisiatif ini.

Efisiensi penggunaan AI: Hanya menggunakan model besar ketika diperlukan, sementara untuk aplikasi sederhana cukup model ringan (lightweight models).

Kolaborasi internasional: Standarisasi regulasi tentang jejak karbon AI, mirip seperti regulasi energi industri.
